```python
import pandas as pd
```

# Tasks

## 1- Read the table from CSV

```python
table = pd.read_csv("/content/Employees.csv")
print(table.info())
table.head(9)
```

## 2- Remove duplicates

```python
table.drop_duplicates(inplace = True)
print(table.info())
table.head(9)
```

## 3- Remove any decimal point in the age coloumn

```python
table["Age"] = (round(table[["Age"]]))
print(table.info())
table.head(9)
```

## 4- convert the USD salary to EGP

```python
table["salary(EGP)"] = table["Salary(USD)"]*30
print(table.info())
table.head(9)
```

## 4- print some stats (AVG age, Median Salaries, Ration "Males/Females", ...)

```python
avg_age = table["Age"].mean()
med_salary = table["salary(EGP)"].median()
num_gender = table["Gender"].value_counts()
t=num_gender["F"]
s=num_gender["M"]
ratio=s/t
print(avg_age,med_salary,num_gender,ratio)
```

## 5- Save the edited table to a CSV file

```python
table.to_csv("/content/newEmployees.csv")
```
